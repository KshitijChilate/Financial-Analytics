```python
# Step1: Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For nicer plots
sns.set(style='whitegrid')
```

```python
# Step 2 : Load Financial Dataset

# Load the data from CSV
df = pd.read_csv("financial_statements.csv")

#Display the first few rows
df.head()
```

```python
#Step 3: Calculate Financial Ratios

#Current Ratio = Current Assets / Current Liabilities
df['Current Ratio'] = df['Current_Assets'] / df['Current_Liabilities']

#Quick Ratio = (Current Assets - Inventory) / Current Liabilities
df['Quick Ratio'] = (df['Current_Assets'] - df['Inventory']) / df['Current_Liabilities']

#Return on Assets (ROA) =Net Income/ Total Assets
df['ROA'] = df['Net_Income'] / df['Total_Assets']

#Return on Equity (ROE) = Net Income / Shareholder's Equity
df['ROE'] = df['Net_Income'] / df['Shareholders_Equity']

#Display updated dataframe with new ratios
df[['Company','Year','Current_Ratio','Quick_Ratio','ROA','ROE']].head()
```

```python
# Step 4: Visualize Ratios Over Years (per company)

#Plot Current Ratio over Years
plt.figure(figsize=(12,6))
sns.lineplot(x='Year',y='Current_Ratio',hue='Company',data=df)
plt.title('Current Ratio Over Years')
plt.ylabel('Current Ratio')
plt.legend()
plt.show()
```

```python
#Plot ROE and ROA

plt.figure(figsize=(12,6))
sns.lineplot(x='Year',y='ROE',hue='Company',data=df ,marker='o',linestyle='--')
sns.lineplot(x='Year',y='ROA',hue='Company',data=df ,marker='s')
plt.title('ROE and ROA Over Years')
plt.ylabel('Ratio')
plt.legend()
plt.show()
```
