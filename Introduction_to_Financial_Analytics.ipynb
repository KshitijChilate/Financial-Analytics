### Step 1: Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import warnings
warnings.filterwarnings('ignore')
```

### Step 3: Load and Priview Dataset

```python
df = pd.read_csv('large_stock_dataset.csv')
df.head()
```

### Step 4:Dataset Structure and Types

```python
df.info()
print("\nShape:",df.shape)
```

### Step 5: Handel the Missing data

```python
print(df.isnull().sum())
df.fillna(method='ffill',inplace=True)
```

### Step 6: Statistical Summary

```python
df.describe()
```

### Step 7:Line Plot for Closing Price

```python
plt.figure(figsize=(14,6))
plt.plot(df['Date'],df['Close'],label='Close',marker='.')
plt.title('Closing Price Over 120 Days')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Step 8:High vs Low Prices

```python
plt.figure(figsize=(14,6))
plt.plot(df['Date'],df['High'],label='High',marker='.')
plt.plot(df['Date'],df['Low'],label='Low',marker='.')
plt.title('High vs Low Prices Over 120 Days')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Step 9: Volume Bar Chart

```python
plt.figure(figsize=(14,5))
plt.bar(df['Date'],df['Volume'],color='purple')
plt.title('Trading Volume Over 120 Days')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

### Step 10: Correlation Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include='number').corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
