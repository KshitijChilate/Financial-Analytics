### Experiment : Enhanced SQL Style Financial Analysis

### Step1 : Create Database

```python
import pandas as pd

# SALES Table
sales_data = {
    'Sale_id': [1, 2, 3, 4],
    'customer_id': [101, 102, 103, 101],
    'item': ['Phone', 'Laptop', 'Phone', 'Tablet'],
    'amount': [30000, 60000, 30000, 15000],
    'profit': [3000, 5000, 4000, 1500],
}

sales_df = pd.DataFrame(sales_data)
print("Sales Data:")
print(sales_df)
```

```python
#CUSTOMER Table
customer_data = {
    'customer_id': [101, 102, 103],
    'customer_name': ['Akash', 'Neha', 'Ramesh'],
    'region':['North','West','East']
}
customers_df = pd.DataFrame(customer_data)
print("\nCustomer Data:")
print(customers_df)
```

### Step 2: Perform INNER JOIN Operation on Sales and Customers

```python
merged_df = pd.merge(sales_df, customers_df, on='customer_id',how='inner')
print("Joined DataFrame:")
print(merged_df)
```

### Step 3: Grouping & Aggregation - Total Sales and Profit per Item

```python
item_group = sales_df.groupby('item')[['amount', 'profit']].sum().reset_index()
print("\nTotal Sales and Profit per Item:")
print(item_group)
```

### Step 4 : Apply HAVING Condition - Filter Items with Profit > 4000

```python
high_profit_items = item_group[item_group['profit'] > 4000]
print("High Profit Items (profit > 4000):")
print(high_profit_items)
```

### Step 5: Region-wise Sales Totals

```python
region_sales = merged_df.groupby('region')[['amount']].sum().reset_index()
print("Region-wise Sales Totals:")
print(region_sales)
```

### Step 6: Visualise Aggregated Financial Insights

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(item_group['item'],item_group['amount'],label='Sales',alpha=1)
plt.bar(item_group['item'],item_group['profit'],label='Profit',alpha=1)
plt.title("Sales and Profit per Item")
plt.xlabel('Item')
plt.ylabel("Amount (INR)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

x= np.arange(len(item_group['item']))
width=0.35

plt.figure(figsize=(8, 5))
plt.bar(x-width/2,item_group['amount'],width,label='Sales',alpha=1,color='skyblue')
plt.bar(x+width/2,item_group['profit'],width,label='Profit',alpha=1,color='orange')

plt.title("Sales and Profit per Item")
plt.xlabel('Item')
plt.ylabel("Amount (INR)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(region_sales['region'], region_sales['amount'], color='skyblue')
plt.title("Region-wise Sales Totals")
plt.xlabel('Region')
plt.ylabel("Total Sales Amount (INR)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
